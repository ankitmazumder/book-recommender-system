```python
from wordcloud import WordCloud, STOPWORDS
def word_cloud(author): 
  comment_words = ''
  stopwords = set(STOPWORDS)
  for val in sub_set.title[(sub_set['authors']==author)& (sub_set['language']!='German')]:
      val = str(val)
      tokens = val.split()
      for i in range(len(tokens)):
          tokens[i] = tokens[i].lower()
     
      comment_words += " ".join(tokens)+" "
 
  wordcloud = WordCloud(width = 400, height = 400,
                  background_color ='white',
                  stopwords = stopwords,
                  min_font_size = 8).generate(comment_words)
                     
  plt.figure(figsize = (5, 5), facecolor = 'k')
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.tight_layout(pad=0)
 
  plt.show()
```

```python
word_cloud('Agatha Christie')
```

```python
word_cloud('Dan Brown')
```

```python
#Feature Engineering
books_cpy.loc[ (books_cpy['average_rating'] >= 0) & (books_cpy['average_rating'] <= 1), 'rating_between'] = "0-1"
books_cpy.loc[ (books_cpy['average_rating'] > 1) & (books_cpy['average_rating'] <= 2), 'rating_between'] = "1-2"
books_cpy.loc[ (books_cpy['average_rating'] > 2) & (books_cpy['average_rating'] <= 3), 'rating_between'] = "2-3"
books_cpy.loc[ (books_cpy['average_rating'] > 3) & (books_cpy['average_rating'] <= 4), 'rating_between'] = "3-4"
books_cpy.loc[ (books_cpy['average_rating'] > 4) & (books_cpy['average_rating'] <= 5), 'rating_between'] = "4-5"
```

```python
books_cpy.head(1)
```

```python
rating_df = pd.get_dummies(books_cpy['rating_between'])
rating_df.head(10)
```

```python
language_df = pd.get_dummies(books_cpy['language'])
language_df.head()
```

```python
features = pd.concat([rating_df, language_df, books_cpy['average_rating'], books_cpy['ratings_count']], axis=1)
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import neighbors
```

```python
min_max_scaler = MinMaxScaler()
features = min_max_scaler.fit_transform(features)
```

```python
model = neighbors.NearestNeighbors(n_neighbors=8, algorithm='auto')
model.fit(features)
dist, idlist = model.kneighbors(features)
```

```python
def Recommender(name):
    book_list = []
    book_id = books_cpy[books_cpy['title'] == name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list.append(books_cpy.loc[newid].title)
    print("Recommended books are:")
    for i in range(0,len(book_list)):
      print(f"{i+1}){book_list[i]}")
```

```python
Recommender('Harry Potter and the Half-Blood Prince (Harry Potter  #6)')
```

```python
def author_Recommender(author):
    auth_list = []
    books=[]
    auth_id = books_cpy[books_cpy['authors'] == author].index
    auth_id = auth_id[0]
    for newid in idlist[auth_id]:
        auth_list.append(books_cpy.loc[newid].authors)
        books.append(books_cpy.loc[newid].title)
    for i in range(0,len(auth_list)):
      print(f"{i+1})Author:{auth_list[i]},\nrecommended book:{books[i]}")
```

```python
author_Recommender('Agatha Christie')
```
